## **1. Experimental Design**



### **What do we want to measure and how to operationalized it?**
As psychologists, we run experiments to **answer research questions** —
to understand how the mind works, what influences behavior, and why
people think and act the way they do.

But a good research question is not enough on its own. To actually
**test** it, we need to translate it into a concrete experimental
procedure. This process is called **operationalization** — turning
abstract concepts into measurable variables.

Every experiment involves at least two key types of variables:

- **Independent variable (IV)**: what we **manipulate**
  (the cause we are testing)
- **Dependent variable (DV)**: what we **measure**
  (the effect we are looking for)

Before writing a single line of code, we always need to ask:

1. What is our **research question**?
2. What is our **IV** — and how do we manipulate it?
3. What is our **DV** — and how do we measure it?
4. How do we **operationalize** each variable concretely?

**A simple example**

Let's start with a simple, everyday example before moving to our
actual experiment.

**Research question:** Does sleep quantity affect mood?

| | Definition | Operationalization |
|---|---|---|
| **IV** | Sleep quantity | Hours of sleep per night (measured with an actigraph) |
| **DV** | Mood | Score on a mood questionnaire (1–10 scale) |

We manipulate (or measure) sleep quantity and observe its effect on mood.
In Python, we might represent this like:

````python
# A simple experiment: does sleep quantity affect mood?

# Independent variable: sleep quantity
# Two conditions: low sleep vs high sleep
sleep_conditions = ["low_sleep", "high_sleep"]  # IV levels

# Dependent variable: mood score (measured after each night)
# This is what we will collect from participants
mood_score = None  # will be filled in during the experiment

# Operationalization:
# low_sleep  = less than 6 hours per night
# high_sleep = more than 8 hours per night
# mood_score = score on a validated mood questionnaire (1-10)

# Each participant will be assigned to one condition:
participant = {
    "id": "P01",
    "condition": "low_sleep",   # ← IV
    "mood_score": None          # ← DV (to be collected!)
}

print(f"Participant {participant['id']} is in the '{participant['condition']}' condition.")
print(f"Mood score: {participant['mood_score']} (not yet collected!)")
````

📌 **Note**: In a real experiment, `mood_score` would not be `None` —
it would be filled in with the actual data collected from the participant.
This is exactly what we will do when we build our experiment!

###**Our task: The Spatial Stroop**

**Why this task?**

To answer a specific research question! Every task we use in psychology
has a reason — it is chosen because it allows us to study a specific
phenomenon in a controlled and measurable way.

One of the most fundamental questions in cognitive psychology is:
**what happens when our brain has to deal with conflicting information?**

For example, when two pieces of information are inconsistent
— one telling us to do one thing, another telling us to do something else
— our brain needs to resolve this conflict. This slows us down and
increases our chances of making errors.

To study this, we need a task that **creates conflict** in a controlled
and measurable way. You have probably heard of the classic **Stroop task**
(word-color version): participants see color words written in a different
ink color (e.g., the word "RED" written in blue ink) and have to name
the ink color while ignoring the word. This creates conflict — and
slows people down!

**A lot of decisions to make!**

Choosing a task is just one of many decisions we need to make when
designing an experiment. Even once we have decided to use the Stroop task,
we still need to decide **which version** to use!

The word-color Stroop requires a **verbal response** (naming
the color out loud), but in a classroom setting, you would all be responding at
the same time, which would be quite chaotic!

Thus, we will use a **spatial version** of the Stroop task, where
both the stimulus and the response are spatial. This means participants
can respond simply by **pressing a key**, making it much easier to
record responses and measure reaction times precisely!





**<font color="red"> The task**

In the **Spatial Stroop task**, participants see an arrow pointing either
**left (←)** or **right (→)**, presented either to the **left** or to
the **right** of a central fixation cross.

The task is simple:
- Respond to the **direction of the arrow** — ignore its position!
- Press **"s"** → arrow points **left ←**
- Press **"l"** → arrow points **right →**






**<font color="red"> Congruency: our independent variable </font>**

The key manipulation is the relationship between the arrow's **direction**
and its **position** on the screen:

| Trial type | Arrow direction | Position | Example |
|---|---|---|---|
| **Congruent** | → right | right side | → presented on the right |
| **Incongruent** | → right | left side | → presented on the left |
| **Congruent** | ← left | left side | ← presented on the left |
| **Incongruent** | ← left | right side | ← presented on the right |

- **Congruent trials**: direction and position **match** → easy!
- **Incongruent trials**: direction and position **conflict** → harder!

Based on what we know about cognitive conflict, we predict:
- **Slower RT** on incongruent trials
- **Lower accuracy** on incongruent trials

This difference (`Incongruent - Congruent`) is called the **Stroop effect**!

Our **<font color="red"> dependent variables </font>** are:
- **Reaction time (RT)**: how long it takes to respond (in ms)
- **Accuracy**: whether the response was correct (1) or incorrect (0)

### **Factorial design**

In psychology experiments, we often want to study **more than one
variable at a time**. A **factorial design** is an experimental design
that includes **two or more independent variables** (called **factors**),
each with two or more **levels**.

The great advantage of factorial designs is that they allow us to study:
- The **main effect** of each factor independently
- The **interaction** between factors — does the effect of one factor
  depend on the level of the other?

Factorial designs are named after the number of levels of each factor:
- **2x2**: 2 factors, each with 2 levels → 4 conditions
- **2x3**: 2 factors, one with 2 levels and one with 3 levels → 6 conditions
- **2x2x2**: 3 factors, each with 2 levels → 8 conditions
- ...and so on!

Today we will focus on the simplest case: a **2x2 factorial design**.

**<font color="red">From one factor to two: building our 2x2 design**

So far in our Spatial Stroop task, we have defined **one factor**:

- **Factor 1: Congruency** (2 levels: congruent vs incongruent)
  → manipulated **between trials** (throughout the experiment, some trials will be congruent
and some will be incongruent)

This allows us to answer our first research question:
*Does cognitive conflict slow people down?*

But now imagine we want to ask a **second research question**:
*Does the eccentricity of the arrow affect the conflict effect?*

In other words — is the Stroop effect larger or smaller when the arrow
is presented further away from the centre of the screen?

To answer this, we add a **second factor**:

- **Factor 2: Eccentricity** (2 levels: central vs peripheral)
  → manipulated **between blocks** (different blocks - slices of the experiment - have different
  eccentricities)

We choose to manipulate eccentricity **between blocks** rather than
between trials because changing the position of the stimulus within
a block could be distracting and confusing for participants.

Now we have a **2x2 factorial design**!

**Our 2x2 design: 4 experimental conditions**

Combining our two factors gives us **4 experimental conditions**:

| | **Central** | **Peripheral** |
|---|---|---|
| **Congruent** | Congruent + Central | Congruent + Peripheral |
| **Incongruent** | Incongruent + Central | Incongruent + Peripheral |

📌 **Note**: Since congruency is manipulated **within blocks** (trial by
trial) and eccentricity is manipulated **between blocks**, we will need:
- **At least, one trial list but we will create different trial lists (e.g., one trial list per block** — each with both congruent and incongruent
  trials, but with a different randomization
- **Two block types**: one central block, one peripheral block
- In total, we will create **multiple trial lists** (one per block),
  all with the same structure but different trial orders!

We will see exactly how to do this in the next section 🐍

It is also correct to create just one trial list and use it in both blocks types.


````python
# Define the levels of each factor
congruency = ["congruent", "incongruent"]   # Factor 1: trial-level
eccentricity = ["central", "peripheral"]    # Factor 2: block-level

# This gives us 4 experimental conditions:
print("Our 2x2 factorial design:")
print("=" * 40) # prints a line of 40 "=" characters — just for visual clarity!
for ecc in eccentricity:
    for cong in congruency:
        print(f"  {ecc} + {cong}")
print("=" * 40)
print(f"Total conditions: {len(congruency) * len(eccentricity)}")
````

### **Experiment structure: trials and blocks**

Another crucial design decision is to define the **structure** of our
experiment — this is essential to ensure that our experiment is designed
in a way that allows us to **effectively answer our research question**!

**<font color="red"> What is a trial? </font>**

A **trial** is the basic unit of an experiment. Each trial represents
one single "event" — in our case, one arrow presentation. A trial
typically follows a fixed sequence of events:

1. **Fixation** → tells the participant to get ready
2. **Stimulus** → the arrow appears
3. **Response** → the participant presses a key
4. **Feedback** (optional) → correct/incorrect

**<font color="red"> What is a block?**

A **block** is a group of consecutive trials. Blocks are useful for:
- Giving participants a **break** during the experiment
- Manipulating a variable **between blocks** (like eccentricity in our case!)
- **Counterbalancing** the order of conditions across the experiment

**<font color="red">How many trials and blocks do we need?**

This is a fundamental question — and the answer always depends on your
**research question**!

A general rule: you need **enough trials and blocks** to:
- Have a reliable estimate of each condition (enough repetitions!)
- Detect the effect you are looking for (statistical power!)
- Not fatigue your participants too much

There is always a trade-off between **more data** (better for statistics)
and **participant fatigue** (shorter is better for them!).



**<font color="red">Our experiment structure**

For our Spatial Stroop experiment, we will use:

````python
# Our experiment structure
n_trials_per_block = 40     # trials per block
n_blocks = 8                # total blocks (4 central + 4 peripheral)
n_trials_total = n_trials_per_block * n_blocks

# Block types
block_types = ["central", "peripheral"]
n_blocks_per_type = n_blocks // len(block_types)  # 4 blocks per type

# Practice block (before the experiment)
n_practice_trials = 10  # short practice block to familiarize participants

print(f"Experiment structure:")
print(f"  Practice block: {n_practice_trials} trials")
print(f"  Experimental blocks: {n_blocks} blocks x {n_trials_per_block} trials")
print(f"  Total experimental trials: {n_trials_total}")
````

## **2. Building the Trial List**

A **trial list** is a list of alle the trials in the experiment. It is a structured table where **each row represents one
trial** and **each column represents one feature** of that trial —
which stimulus to show, which condition it belongs to, what the correct
response is, and so on.

Think of it as a recipe book for your experiment: Python (and PsychoPy)
will go through it row by row, and for each trial, they will know
exactly what to do!



````python
# A single trial in our experiment — represented as a dictionary
trial = {
    "trial_number":     1,              # position in the block
    "block_number":     1,              # position in the experiment
    "direction":        "right",        # arrow direction (right or left)
    "position":         "right",        # arrow position on screen (right or left)
    "congruency":       "congruent",    # congruent or incongruent
    "correct_response": "l",            # "l" for right, "s" for left
    "eccentricity":     "central"       # central or peripheral (block-level)
}

print("Example trial:")
for key, value in trial.items():
    print(f"  {key}: {value}")
````

A complete **trial list** is simply a collection of many such trials —
one for each trial in the experiment!

**<font color="red"> What makes a good trial list? </font>**

Not all trial lists are equal! A good trial list must satisfy
a few important requirements:

**1. Balanced conditions**

Each experimental condition should appear the **same number of times**
(unless you have a specific reason not to — e.g., go/no-go tasks).
This ensures that your results are not biased by one condition
appearing more often than others.

````python
# ✅ Balanced: each condition appears the same number of times
balanced = ["congruent", "incongruent", "congruent", "incongruent",
            "congruent", "incongruent", "congruent", "incongruent"]

# ❌ Unbalanced: congruent appears much more often!
unbalanced = ["congruent", "congruent", "congruent", "congruent",
              "congruent", "congruent", "incongruent", "incongruent"]

# Check balance
from collections import Counter
print(f"Balanced: {Counter(balanced)}")
print(f"Unbalanced: {Counter(unbalanced)}")
````

**2. Randomized trial order**

Trials should be presented in a **random order** to avoid systematic
effects — for example, participants getting better simply because they
learn the pattern of the stimuli, or getting worse because of fatigue
at the end.

````python
import random

conditions = ["congruent", "incongruent"] * 4  # 4 of each
print(f"Before randomization: {conditions}")

random.shuffle(conditions)
print(f"After randomization:  {conditions}")
````

**3. No problematic sequences**

Even with randomization, sometimes we want to **control** what
sequences of trials can occur. For example, we might want to avoid
having more than 3 trials of the same condition in a row —
this could make the task too predictable or too easy!

````python
# Example of a problematic sequence — too many incongruent in a row!
problematic = ["congruent", "incongruent", "incongruent", "incongruent",
               "incongruent", "incongruent", "congruent", "congruent"]

# Check for runs of more than 3 consecutive same condition
max_run = 1
current_run = 1
for i in range(1, len(problematic)):
    if problematic[i] == problematic[i-1]:
        current_run += 1
        max_run = max(max_run, current_run)
    else:
        current_run = 1

print(f"Longest run of same condition: {max_run} trials")
if max_run > 3:
    print("⚠️ Warning: sequence has runs longer than 3 — consider re-randomizing!")
else:
    print("✅ Sequence looks good!")
````

**<font color="red"> How we will build our trial list: step by step </font>**

We will build our trial list following these steps:

1. **Define trial features and create all possible combinations**
2. **Add repetitions**
3. **Randomize** (with criteria!)
4. **Export as CSV** → ready to import in PsychoPy!

Let's go through them one by one!

### **Step 1: Define the trial-features and create all possible combinations**


**<font color="red"> Where do we start? </font>**

When building a trial list, a good rule of thumb is to always start
from the **simplest factor** — and specifically from the factors that
are manipulated **at the trial level** (i.e., that change from trial
to trial within a block).

In our experiment:
- **Factor 1: Congruency** → manipulated at the **trial level**
  ✅ starts here!
- **Factor 2: Eccentricity** → manipulated at the **block level**
  ❌ does NOT go in the trial list base!

If we had more than one factor at the trial level, the logic would be:
- Start from **one factor** → get the base combinations
- Then **multiply** by the levels of the second factor → double the combinations
- Then **multiply** again for a third factor → and so on!

But for now, we only have one trial-level factor, so let's start there!

**<font color="red"> The building blocks of our trial list </font>**

Congruency **emerges** from
the combination of two fundamental stimulus features:

- **Direction**: the direction the arrow points → right (→) or left (←)
- **Position**: where the arrow appears on the screen → right or left

From these two features — each with 2 levels — we get
**4 possible combinations**:

| Direction | Position | Congruency | Correct response |
|---|---|---|---|
| right (→) | right | congruent | "l" |
| right (→) | left | incongruent | "l" |
| left (←) | left | congruent | "s" |
| left (←) | right | incongruent | "s" |

Two important things to notice:
- **Congruency** is a **consequence** of the combination of direction
  and position — not an independent variable we set directly!
- **Correct response** is derived from **direction only**
  (not position!) — because the task is to respond to where the arrow
  **points**, ignoring where it **appears**!

**<font color="red"> What about Factor 2: Eccentricity? </font>**

Eccentricity is manipulated **between blocks** — meaning that within
a single block, all trials have the **same eccentricity**.
This is why it does not belong in our trial list base:
it would make no sense to have trials with different eccentricities
within the same block, as this would defeat the purpose of the
block-level manipulation!

Instead, we will create **different block types**:
- A **central block** → all trials use central eccentricity
- A **peripheral block** → all trials use peripheral eccentricity

The trial list base (with congruent and incongruent trials) will be
the same for both block types — only the eccentricity parameter
will change at the block level.

📌 **Note**: If we wanted to manipulate eccentricity at the **trial
level** instead, we would simply **multiply** our base trial list by
the levels of eccentricity (central, peripheral) — going from
4 to 8 combinations. But this would mean eccentricity changes
from trial to trial within a block, which is not what we want here!

#### Let's build it in Python!


````python
# --------------------------------------------------------------
# STEP 1: nested loop — to create the four basic combinations
# --------------------------------------------------------------

# Define the possible levels of each trial feature
directions = ["right", "left"]   # arrow direction
positions  = ["right", "left"]   # arrow position on screen

# Create all possible combinations using a nested loop
# For each combination, we also derive:
# - congruency (from direction and position)
# - correct_response (from direction only!)
trial_list_base = []  # empty list — we will fill it trial by trial

for direction in directions:
    for position in positions:

        # Step 1a: determine congruency
        # congruent = direction and position match
        # incongruent = direction and position do not match
        if direction == position:
            congruency = "congruent"
        else:
            congruency = "incongruent"

        # Step 1b: determine correct response
        # correct response is ALWAYS derived from direction!
        # "l" for right, "s" for left
        if direction == "right":
            correct_response = "l"
        else:
            correct_response = "s"

        # Step 1c: store this trial as a dictionary
        trial = {
            "direction":        direction,
            "position":         position,
            "congruency":       congruency,
            "correct_response": correct_response
        }

        # Step 1d: add to the trial list
        trial_list_base.append(trial)

# Inspect the result
print(f"Number of unique combinations: {len(trial_list_base)}")
print()
print(f"{'direction':<10} {'position':<10} {'congruency':<15} {'correct_resp':<10}")
print("-" * 50)
for trial in trial_list_base:
    print(f"{trial['direction']:<10} {trial['position']:<10} "
          f"{trial['congruency']:<15} {trial['correct_response']:<10}")
````

📌 **Note**: In this example we used a nested loop, which is great
for understanding the logic. In practice, when you have many factors
or many levels, you might find `itertools.product()` more convenient
— it generates all combinations automatically in one line!
Feel free to explore it in the documentation 😊

We now have our **base trial list** with 4 unique combinations —
the building block of our experiment!

Notice that:
- We have **2 congruent** trials (right+right, left+left)
  and **2 incongruent** trials (right+left, left+right) → **balanced!** ✅
- The correct response always follows the direction, never the position ✅
- Eccentricity is not here yet — it will be added at the block level ✅

But 4 trials is not enough for a real experiment!
In the next step, we will add **repetitions** to reach our
target of 40 trials per block.

### **Step 2: Add repetitions**

**<font color="red"> Why do we need repetitions? </font>**

4 trials is definitely not enough for a real experiment! We need to
repeat each condition multiple times for two main reasons:

1. **Reliable estimates**: with only a few trials per condition,
   a single slow or fast response could completely distort our results.
   More repetitions → more stable and reliable estimates of RT and accuracy!

2. **Statistical power**: the more trials we have, the better our
   chances of detecting a real effect if it exists. This is what
   statisticians call **statistical power** — we will come back to
   this concept when we analyze our data!

There is however always a trade-off: **more trials = more fatigue**
for the participant. So we need to find the right balance!

**<font color="red"> How many repetitions do we need? </font>**

📌 **General rule of thumb**: aim for a **minimum of ~30 trials per condition**
across the whole experiment!

Let's apply this to our experiment.
In our 2x2 design, we have **4 conditions**:
- Congruent + Central
- Congruent + Peripheral
- Incongruent + Central
- Incongruent + Peripheral

Let's count how many trials per condition we get with our design:
- Each block contains **40 trials**: 20 congruent + 20 incongruent
- We have **4 central blocks** → 80 congruent+central
  and 80 incongruent+central trials
- We have **4 peripheral blocks** → 80 congruent+peripheral
  and 80 incongruent+peripheral trials

→ **80 trials per condition** ✅ well above the minimum of 30!

We chose 40 trials per block (10 repetitions x 4 combinations)
because it gives us enough data while keeping each block
short enough to avoid excessive fatigue.

📌 **Note**: This is a relatively simple experiment! As you design
more complex experiments with more conditions and manipulations,
reaching the minimum number of trials per condition becomes
increasingly challenging. But remember: **more is always better** —
as long as your participants can handle it! 😊

````python
# ========================
# STEP 2: Add repetitions
# ========================

# Define number of repetitions per combination
n_repetitions = 10   # 4 combinations x 10 = 40 trials per block

# Add repetitions
# ⚠️ Important: we use .copy() for each trial to avoid reference problems!
# Why? Because dictionaries in Python are stored by REFERENCE,
# not by value. This means that without .copy(), all repetitions
# of the same trial would point to the SAME dictionary in memory —
# if you modify one, you modify all of them!
#
# Example of the problem WITHOUT .copy():
# trial_list = [trial_list_base] * 10  ← all point to same object!
# trial_list[0]["direction"] = "up"    ← changes ALL trials!
#
# With .copy(), each trial is an INDEPENDENT copy → safe to modify!

trial_list = []
for _ in range(n_repetitions):
    for trial in trial_list_base:
        trial_list.append(trial.copy())

print(f"Trial list after repetitions: {len(trial_list)} trials")
print(f"Expected: {len(trial_list_base)} combinations x "
      f"{n_repetitions} repetitions = "
      f"{len(trial_list_base) * n_repetitions} trials")

# Check balance — should be perfectly balanced!
from collections import Counter

congruency_counts = Counter([t["congruency"] for t in trial_list])
direction_counts  = Counter([t["direction"]  for t in trial_list])
position_counts   = Counter([t["position"]   for t in trial_list])

print(f"\nBalance check:")
print(f"  Congruency: {dict(congruency_counts)}")
print(f"  Direction:  {dict(direction_counts)}")
print(f"  Position:   {dict(position_counts)}")
````

📌 **Note on balance**: as you can see, our trial list is perfectly
balanced — each combination appears exactly the same number of times.

This is not a coincidence! It is a direct consequence of starting
from a **balanced base** (4 combinations, 2 congruent and 2 incongruent)
and simply repeating it.

This is exactly why it is always a good idea to start from the
**minimal balanced unit** — once you get the base right,
balance is guaranteed no matter how many repetitions you add! ✅

### **Step 3: Randomize**

**<font color="red"> Why do we need to randomize? </font>**

Having a balanced trial list is not enough — we also need to present
the trials in a **random order**!

Without randomization, participants could:
- **Learn the pattern** of stimuli and start anticipating responses
- Show **fatigue effects** that are confounded with condition
  (e.g., if all incongruent trials come at the end)
- Show **practice effects** that systematically affect one condition
  more than another

Randomization ensures that any such effects are **spread evenly**
across all conditions!

**<font color="red"> The naive approach </font>**

The simplest way to randomize is to use `random.shuffle()` —
but as we will see, this is not always enough!

````python
import random

# Make a copy to not mess up our trial_list!
trial_list_naive = trial_list.copy()
random.shuffle(trial_list_naive)

# Let's look at the first 10 trials...
print("Trials after basic shuffle:")
print(f"{'direction':<10} {'position':<10} {'congruency':<15}")
print("-" * 40)
for trial in trial_list_naive[:10]:
    print(f"{trial['direction']:<10} {trial['position']:<10} "
          f"{trial['congruency']:<15}")
````

Run the cell above a few times — you might get something like this:

| Trial | Direction | Position | Congruency |
|---|---|---|---|
| 1 | right | right | **congruent** |
| 2 | left | left | **congruent** |
| 3 | right | right | **congruent** |
| 4 | left | left | **congruent** |
| 5 | left | left | **congruent** |
| ... | ... | ... | ... |

⚠️ **5 congruent trials in a row!** This is a problem:
- Participants might **anticipate** the next trial type
- The task becomes temporarily **too easy** (or too hard)
- This could **bias our RT estimates** for those trials

`random.shuffle()` has no memory — it does not care about what
came before! We need a smarter approach.

**<font color="red"> Randomization with constraints </font>**

A better approach is to randomize while **controlling for problematic
sequences**. In our experiment, we want to avoid more than
**3 consecutive trials** with the same:

- **Congruency** (congruent or incongruent)
  → avoid blocks of very easy or very hard trials

- **Direction** (right or left)
  → avoid participants developing a response bias

- **Position** (right or left)
  → avoid spatial attention being captured by one side

- **Stimulus** (direction + position combination)
  → avoid exact same stimulus appearing too many times in a row

📌 **Why max 3?** In our task we only have 2 options for each feature
(e.g., right or left). With so few options, it is mathematically very
hard to avoid any repetition — so we allow up to 3 consecutive repeats.
If you had 4 stimuli or 4 response options, you could set a stricter
criterion (e.g., max 2)!

To implement this, we will use a **re-randomization approach**:
1. Shuffle the trial list
2. Check if all criteria are satisfied
3. If not → shuffle again and check again
4. Repeat until a valid sequence is found (or until we reach a
   maximum number of attempts!)

````python
# ========================================
# STEP 3a: Define randomization functions
# ========================================


def check_criteria(trial_list, max_consecutive=3):
    """
    Check if a trial list satisfies ALL randomization criteria:
    - No more than max_consecutive same congruency in a row
    - No more than max_consecutive same direction in a row
    - No more than max_consecutive same position in a row
    - No more than max_consecutive same stimulus (dir+pos) in a row

    Returns True if the sequence is valid, False otherwise.
    """
    # We slide a "window" of (max_consecutive + 1) trials
    # across the trial list, checking one position at a time.
    # If ALL trials in the window are the same → criterion violated!
    for i in range(max_consecutive, len(trial_list)):

        # Current window: from i-max_consecutive to i (included)
        # e.g., with max_consecutive=3: window has 4 trials
        window = trial_list[i - max_consecutive:i + 1]

        # Check congruency: are all trials in the window congruent/incongruent?
        # set() removes duplicates → if len==1, all values are the same!
        if len(set([t["congruency"] for t in window])) == 1:
            return False  # ❌ criterion violated!

        # Check direction
        if len(set([t["direction"] for t in window])) == 1:
            return False  # ❌ criterion violated!

        # Check position
        if len(set([t["position"] for t in window])) == 1:
            return False  # ❌ criterion violated!

        # Check stimulus (direction + position combined)
        # e.g., "right_right", "right_left", etc.
        if len(set([f"{t['direction']}_{t['position']}"
                    for t in window])) == 1:
            return False  # ❌ criterion violated!

    return True   # ✅ all criteria satisfied!


def count_violations(trial_list, max_consecutive=3):
    """
    Count the total number of criterion violations in a trial list.
    Used to keep track of the BEST sequence found so far,
    in case no perfect sequence is found within max_attempts.
    """
    violations = 0
    for i in range(max_consecutive, len(trial_list)):
        window = trial_list[i - max_consecutive:i + 1]

        # Each violated criterion counts as 1 violation
        if len(set([t["congruency"] for t in window])) == 1:
            violations += 1
        if len(set([t["direction"] for t in window])) == 1:
            violations += 1
        if len(set([t["position"] for t in window])) == 1:
            violations += 1
        if len(set([f"{t['direction']}_{t['position']}"
                    for t in window])) == 1:
            violations += 1

    return violations   # lower is better!


def randomize_with_criteria(trial_list, max_consecutive=3, max_attempts=1000):
    """
    Randomize a trial list while satisfying all criteria.

    Strategy:
    - Shuffle → check → if valid: return ✅
    - If not valid → shuffle again → check again
    - After max_attempts: return the BEST sequence found
      (fewest violations) with a warning ⚠️
    """
    best_sequence = None  # best sequence found so far
    best_score = -1       # -1 means "not yet initialized"

    for attempt in range(max_attempts):

        # Shuffle the trial list
        random.shuffle(trial_list)

        # Check if all criteria are satisfied
        if check_criteria(trial_list, max_consecutive):
            print(f"✅ Valid sequence found after {attempt + 1} attempts!")
            return trial_list # ← return immediately if valid!

        # If not valid, count violations and update best sequence
        score = count_violations(trial_list, max_consecutive)
        if best_score == -1 or score < best_score:
            best_score = score
            best_sequence = trial_list.copy() # save best so far

    # If we get here, no perfect sequence was found
    print(f"⚠️ Warning: no valid sequence found after {max_attempts} attempts!")
    print(f"   Using best sequence found ({best_score} violations)")
    return best_sequence
````

````python
# =================================
# STEP 3b: Randomize with criteria!
# =================================

trial_list_randomized = randomize_with_criteria(
    trial_list,
    max_consecutive=3, # can be decreased if you have more stimuli/responses/or
                       # increased if no valid solution can be found
    max_attempts=10000  # increase if you have trouble finding a valid sequence
)

# Quick check — let's look at the first 10 trial
print(f"\nFirst 10 trials after randomization:")
print(f"{'direction':<10} {'position':<10} {'congruency':<15} {'correct':<10}")
print("-" * 50)
for trial in trial_list_randomized[:10]:
    print(f"{trial['direction']:<10} {trial['position']:<10} "
          f"{trial['congruency']:<15} {trial['correct_response']:<10}")
````

### **Creating multiple trial lists**

We now have our first trial list 🥳 — `trial_list_1.csv`.

We could stop here and use this same trial list for all 8 blocks.
However, this is **not ideal**:
- Participants might start **recognizing the sequence** after a few blocks
- Any **order effects** specific to this particular randomization
  would repeat across all blocks

Instead, we will create **4 different trial lists** — each with a
different randomization — and then use a clever trick to get 4 more!

To create trial lists 2, 3 and 4, we simply need to **repeat Steps 3
and 4** with a different randomization each time. Instead of doing
this manually three times, we can use a **loop** to automate it!

📌 **Note**: Steps 1 and 2 stay the same — we always start from the
same base combinations and repetitions. Only the randomization
(Step 3) and the filename (Step 4) change!

````python
# Create trial lists 2, 3 and 4
# (trial_list_1 was already created above!)

for i in range(2, 5):  # i = 2, 3, 4

    # Step 3: randomize with criteria
    # ⚠️ Important: use trial_list.copy() to start fresh each time!
    # randomize_with_criteria() modifies the list in place with shuffle()
    # so without .copy() the next iteration would start from an
    # already-shuffled list!
    trial_list_temp = trial_list.copy()
    trial_list_new = randomize_with_criteria(
        trial_list_temp,
        max_consecutive=3,
        max_attempts=1000
    )

    # Add trial numbers — starting from 1 for each trial list!
    for j, trial in enumerate(trial_list_new):
        trial["trial_number"] = j + 1

    # Step 4: export as CSV
    df = pd.DataFrame(trial_list_new)
    df = df[["trial_number", "direction", "position",
             "congruency", "correct_response"]]

    filename = f"trial_list_{i}.csv"
    df.to_csv(filename, index=False)
    print(f"✅ Saved {filename}")

print(f"\n✅ Done! We now have 4 trial lists: "
      f"trial_list_1.csv to trial_list_4.csv")
````

### **A clever trick: creating 4 more trial lists by reversing!**

We now have 4 different trial lists. But we need 8 — one per block!

Instead of creating 4 more from scratch, we can use a clever trick:
**reverse the order of the existing trial lists**!

A reversed trial list:
- Has the **same trials** → still balanced ✅
- Has a **different order** → not the same as the original ✅
- Still satisfies our **randomization criteria** → we will check! ✅

This gives us 4 more trial lists for free — without any additional
randomization needed!

To do this, we will:
1. **Reload** the 4 trial lists from the CSV files we just saved
2. **Reverse** the order of each one
3. **Update the trial numbers** (so they start from 1 again!)
4. **Save** the reversed trial lists as trial_list_5 to trial_list_8

````python
import pandas as pd

# Reload the 4 trial lists from CSV
# and create 4 reversed versions

for i in range(1, 5):  # i = 1, 2, 3, 4

    # Step 1: reload from CSV
    df = pd.read_csv(f"trial_list_{i}.csv")
    print(f"Loaded trial_list_{i}.csv — shape: {df.shape}")

    # Step 2: reverse the order
    # .iloc[::-1] reverses the rows of the DataFrame
    # reset_index(drop=True) resets the index after reversing
    df_reversed = df.iloc[::-1].reset_index(drop=True)

    # Step 3: update trial numbers — starting from 1 again!
    df_reversed["trial_number"] = range(1, len(df_reversed) + 1)

    # Reorder columns nicely
    df_reversed = df_reversed[["trial_number", "direction", "position",
                                "congruency", "correct_response"]]

    # Step 4: save as CSV
    filename = f"trial_list_{i + 4}.csv"
    df_reversed.to_csv(filename, index=False)
    print(f"✅ Saved {filename}")
    print()

print("✅ Done! We now have 8 trial lists:")
for i in range(1, 9):
    print(f"  trial_list_{i}.csv")
````

We now have **8 different trial lists** — each balanced, randomized,
and with trial numbers starting from 1! 🎉

But there is one last thing we need to decide...


### **Block order**

Having 8 blocks is not enough — we also need to decide the **order**
in which they are presented!

A naive approach would be to present all central blocks first,
then all peripheral blocks:

❌ **Bad order**: central, central, central, central,
                  peripheral, peripheral, peripheral, peripheral

This is problematic because:
- Any **practice effects** (getting better over time) would
  systematically favor the second condition (peripheral)
- Any **fatigue effects** would systematically disadvantage
  the second condition
- Participants might **notice the change** halfway through
  and adjust their strategy

A much better approach is to **alternate and counterbalance**
the block types — so that both conditions appear roughly equally
often at the beginning, middle and end of the experiment!

**<font color="red"> Our block order: ABBABAAB </font>**

We will use the following counterbalanced order
(A = central, B = peripheral):

| Block | 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 |
|---|---|---|---|---|---|---|---|---|
| **Type** | A | B | B | A | B | A | A | B |
| | central | peripheral | peripheral | central | peripheral | central | central | peripheral |

This order ensures that:
- Each condition appears **4 times** → balanced ✅
- Neither condition is always first or always last ✅
- The order alternates regularly → practice and fatigue
  effects are spread evenly ✅

**Other valid counterbalanced orders**

ABBABAAB is just one of many valid options!
Here are some alternatives:

````python
# Some examples of valid counterbalanced orders
# (4 central + 4 peripheral, alternating)

order_1 = ["C", "P", "P", "C", "P", "C", "C", "P"]  # ABBABAAB (ours!)
order_2 = ["P", "C", "C", "P", "C", "P", "P", "C"]  # BAABABBA (reversed!)
order_3 = ["C", "P", "C", "P", "P", "C", "P", "C"]  # ABABABBA
order_4 = ["P", "C", "P", "C", "P", "C", "C", "P"]  # BABABAAB

# All have 4 C and 4 P!
for name, order in [("Order 1 (ours)", order_1),
                     ("Order 2", order_2),
                     ("Order 3", order_3),
                     ("Order 4", order_4)]:
    print(f"{name}: {order}")
    print(f"  Central blocks: {order.count('C')} — "
          f"Peripheral blocks: {order.count('P')}")
    print()
````

**<font color="red"> Assigning trial lists to blocks </font>**

Now we can assign our 8 trial lists to the 8 blocks following
our counterbalanced order!

For convenience, it is useful to add two extra columns to each
trial list:
- **block_number** → which block this trial list belongs to (1-8)
- **block_type** → central or peripheral

This way, when PsychoPy loads the trial list, it already knows
everything it needs about each trial — including which block
it belongs to and what eccentricity to use!

````python
import pandas as pd

# Our counterbalanced block order
block_order = ["central", "peripheral", "peripheral", "central",
               "peripheral", "central", "central", "peripheral"]

# Assign trial lists to blocks and add block info
for block_num, block_type in enumerate(block_order):

    trial_list_num = block_num + 1  # trial_list_1 → block 1, etc.

    # Reload the trial list
    df = pd.read_csv(f"trial_list_{trial_list_num}.csv")

    # Add block information
    df.insert(1, "block_number", block_num + 1)   # after trial_number
    df.insert(2, "block_type", block_type)         # after block_number

    # Save updated trial list
    filename = f"trial_list_{trial_list_num}.csv"
    df.to_csv(filename, index=False)

    print(f"✅ trial_list_{trial_list_num}.csv → "
          f"Block {block_num + 1} ({block_type})")

print(f"\n✅ All 8 trial lists updated with block information!")
print(f"\nFinal structure of each trial list:")
print(f"  Columns: trial_number, block_number, block_type, "
      f"direction, position, congruency, correct_response")
print(f"  Rows: 40 trials per block")
````

##**3. Block and trial loops, and STIMULUS presentation**

    arrow.image = 'stim/arrow_' + trial_list['direction'][trial]
    arrow.pos   = POSITION[trial_list['position'][trial]]
    fixation.draw()  # Prepare (fixation cross)...

Now we're ready to collect the **PARTICIPANT'S RESPONSE!** We'll use the psychopy module  **`event`** to control the inputs that we get with the keyboard.

First, we must prepare the enviroment, specifically:
- We use the function `event.clearEvents()` to clear any leftover events from our keyboard (i.e., keypresses performed before the stimulus onset).
- We also reset our stopwatch (`clock`) so it starts exactly when the stimulus appears.

Then, the function **`event.waitKeys()`** will pause the script and listen for a key press. We'll give it three important parameters:
- the maximum time window that it should wait (`maxWait`), here, 2 seconds.
- a list of allowed keys presses (`keyList`), here, 's' and 'l'.
- and a reference to compute the keypress RT (`timeStamped`), here, our `clock`.

The function **output** is a list of `[key, rt]` pairs (*tuple*) — or `None` if no key was pressed in time. We then unpack that into three variables: the key pressed, the reaction time, and whether the response was correct.

```python
    # clear past keyboard events
    event.clearEvents()
    # Restart the clock with the stimulus (onset = 0ms)
    clock.reset()

    # Collect key presses for 2 secons     
    keys = event.waitKeys(
        maxWait=2.0,
        keyList= keys_allowed,
        timeStamped=clock
    )
    # Proccess the responses to get: key pressed, RT, accuracy
    if keys: # If a key has been pressed...
        key, rt = keys[0]
        correct = int(key == trial_list['correct_response'][trial])
    else:
        # Trials with no response
        key, rt, correct = 'none', None, 0  
```
